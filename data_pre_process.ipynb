```python
adj=pd.read_table("caculate1.csv", sep=",",encoding='utf8')
adj_np = np.array(adj) #先将数据框转换为数组
adj_list = adj_np.tolist()  #其次转换为列表
adj_dic = {}
for i in adj_list:
    adj_dic[i[0]] = i[1:]
    
```

```python
n = 0
for sheet_name in sheet_na:
    n += 1
    #print(sheet_name)
    _df = pd.read_excel(path + xlsx_names1, sheet_name=sheet_name,header=None)
    if np.isnan(adj_dic[sheet_name][1]):
        AA = _df[8].loc[7:]*adj_dic[sheet_name][0]*10/(_df[9].loc[7:]*4)
        #print(_df[8].loc[7:]/_df[9].loc[7:])
        train_data = np.array(AA) #先将数据框转换为数组
        train_data_list = train_data.tolist()  #其次转换为列表
        #print(np.array(train_data_list))  #以数组形式打出来方便看
    else:
        AA = _df[8].loc[7:]*adj_dic[sheet_name][0]*10/(_df[9].loc[7:]*4*adj_dic[sheet_name][1])
        train_data = np.array(AA) #先将数据框转换为数组
        train_data_list = train_data.tolist()  #其次转换为列表
    if n ==1:
        df = _df.iloc[6:,0:6]
        df.columns = df.iloc[0]
        df = df.drop([6])
    df[sheet_name]= train_data_list
df.to_csv("File_result.csv")
#print(df)
```
